```python
from apify_client import ApifyClient
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
client = ApifyClient(os.getenv("apify_key"))

# ArXiv scrape
run = client.actor("fatihai-tools/arxiv-scraper").call(run_input={
    "maxItems": 20,
    "searchQuery": "large language models"
})
```

```python
run = client.actor("primeparse/google-scholar-scraper").call(run_input={
    "query": "Generative AI",
    "maxPages": 1,
    "startYear": 2022,
    "endYear": 2026,
    "proxyConfiguration": {
    "useApifyProxy": True
  }
})
```

```python
results = [item for item in client.dataset(run.default_dataset_id).iterate_items()]
```
