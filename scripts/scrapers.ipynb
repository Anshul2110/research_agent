```python
from apify_client import ApifyClient
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
client = ApifyClient(os.getenv("apify_key"))

# ArXiv scrape
run1 = client.actor("fatihai-tools/arxiv-scraper").call(run_input={
    "maxItems": 2,
    "searchQuery": "large language models"
})
```

```python
## google scholar scrape
run2 = client.actor("primeparse/google-scholar-scraper").call(run_input={
    "query": "Generative AI",
    "maxPages": 1,
    "startYear": 2022,
    "endYear": 2026,
    "proxyConfiguration": {
    "useApifyProxy": True
  }
})
```

```python
## semantic scholar scrape 
run3 = client.actor("agenscrape/semantic-scholar-paper-scraper").call(run_input={


    "keyword": "generative AI",
    "maxResults": 5


})
```

```python
for item in client.dataset(run1.default_dataset_id).iterate_items():
    print("ArXiv item:", item)
```

```python
res1 = []
for item in client.dataset(run1.default_dataset_id).iterate_items():
    res1.append({
        "title": item["title"],
        "abstract": item["abstract"],
        "authors": item["authors"],
        "publishedYear": item["published"],
        "url": item["url"],
    })

res2 = []
for item in client.dataset(run2.default_dataset_id).iterate_items():
    res2.append({
        "title": item["title"],
        "abstract": item["abstract"],
        "authors": item["authors"],
        "publishedYear": item["year"],
        "url": item["scholarLink"],
    })

res3 = []
for item in client.dataset(run3.default_dataset_id).iterate_items():
    res3.append({
        "title": item["title"],
        "abstract": item["abstract"],
        "authors": item["authors"],
        "publishedYear": item["year"],
        "url": item["url"],
    })

```

```python
results = res1 + res2 + res3
```

```python
len(results)
```

```python
with open("results.json", "w") as f:
    import json
    json.dump(results, f, indent=4)
```
